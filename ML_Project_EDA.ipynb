```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Reading the csv datasetets and getting some information about them.
df_high = pd.read_csv("high_popularity_spotify_data.csv")
df_high
```

```python
df_high.info()
```

```python
df_low = pd.read_csv("low_popularity_spotify_data.csv")
df_low
```

```python
df_low.info()
```

```python
# Concatenating those two datasets to have 1 and to make the work on them easier.
#Creating new column for summarizing from which dataset the rows are coming from.
df_low["pop_class"] = 0
df_high["pop_class"] = 1
df = pd.concat([df_low, df_high],ignore_index=True)
df
```

```python
#Dropping columns that couldn't help during the tasks like the columns containing links, ids, etc.
drop_cols = [
    "track_href","uri","analysis_url","id",
    "track_album_id","playlist_id","track_id"
]
df = df.drop(columns=drop_cols)
df
```

```python
# Select only numerical columns from the dataset.
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

```python
# Plot histograms to inspect the distribution of all numerical features.
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
plt.show()
```

## Distribution of Audio Features

The histograms show that the Spotify audio features have very different scales and distribution shapes, which justifies the later use of StandardScaler.

- Energy, danceability, valence follow roughly uniform-to-normal distributions centered in the mid-range.  
- Tempo is moderately skewed, with most songs between 90–140 BPM.  
- Duration_ms is right-skewed, with many short songs and a long tail of longer tracks.  
- Acousticness and instrumentalness show very strong right-skewness.  

**These inconsistencies in scale and skewness confirm that scaling and PCA are necessary before clustering.**

```python
# Plotting the correlation heatmap for numerical columns (features).
num_df = df.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(14,10))
sns.heatmap(num_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

## Correlation Structure Between Features

The heatmap reveals several meaningful relationships:

- Energy, loudness, and danceability are strongly positively correlated, suggesting that energetic and louder songs tend to be more danceable.
- Acousticness is strongly negatively correlated with energy and loudness, which reflects the difference between acoustic and electronic/produced music.
- etc.

**These correlations help explain why certain genres may overlap in the PCA plot and why clusters may form around shared musical traits such as energy or acousticness.**

```python
features = df.drop([
    "track_album_name", "track_artist", "track_name",
    "playlist_name","playlist_genre","playlist_subgenre",
    "type","track_album_release_date"
], axis=1)

for f in features:
    plt.figure(figsize=(8,4))
    sns.kdeplot(data=df, x=f, hue="pop_class", fill=True)
    plt.title(f)
    plt.show()
```

## Interpretation of Boxplots by Genre

Together, the boxplots indicate that:
- Genres differ systematically in acousticness, energy, tempo, and duration.
- Certain genres are highly distinct (e.g., classical, ambient, rock), while others overlap heavily (e.g., pop, k-pop arabic pop or afrobeats and latin, or arabic and turkish).
- Here the strongest discriminators appear to be acousticness, energy, and duration, while tempo is more mixed.

**These clear contrasts across genres justify the use of clustering, because the features exhibit structure that unsupervised learning methods can attempt to capture.**
